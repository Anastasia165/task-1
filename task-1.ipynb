```python
import pandas as pd
import numpy as np
```

```python
# Загрузка таблиц 
table_plus = pd.read_csv('plus_41718.csv')  
table_minus = pd.read_csv('minus_41718.csv') 

table_plus['Процент'] = (table_plus['Процент'] * 100).round(2)
table_minus['Процент'] = (table_minus['Процент'] * 100).round(2)

# styled_df = table_plus.style.format({
#     'Процент': '{:.2%}'  # Отображает как проценты, но данные остаются дробями
# })
```

```python
print("Исходная таблица 1:")
display(table_plus.head())
print("\nИсходная таблица 2:")
display(table_minus.head())
```

```python
# 1. Таблицы с одинаковыми ключевыми словами (внутреннее соединение)
# Используем merge для нахождения пересечений по ключевым словам
common_keywords = pd.merge(table_plus, table_minus, on='Топик', how='inner', suffixes=('_table_plus', '_table_minus'))

pd.set_option('display.width', 1000)  # максимальная ширина
pd.set_option('display.max_columns', None)  # показывать все колонки
pd.set_option('display.expand_frame_repr', False)  # не переносить на новую строку

print("\n" + "="*80)
print("1. ОБЩИЕ КЛЮЧЕВЫЕ СЛОВА (пересечение):")
print("="*80)
print("Количество общих ключевых слов:", len(common_keywords['Топик'].unique()))
print("\nТаблица с общими ключевыми словами и значениями из обеих таблиц:")
display(common_keywords)
```

```python
print("\n" + "="*80)
print("2. РАЗЛИЧИЯ:")
print("="*80)

# 2. Различия - ключевые слова, которые есть только в первой таблице
only_in_table_plus = pd.merge(table_plus, table_minus[['Топик']], on='Топик', how='left', indicator=True)
only_in_table_plus = only_in_table_plus[only_in_table_plus['_merge'] == 'left_only'].drop('_merge', axis=1)

print("\nA) Ключевые слова, которые есть ТОЛЬКО в первой таблице:")
print(f"Количество: {len(only_in_table_plus)}")
if len(only_in_table_plus) > 0:
    display(only_in_table_plus)
else:
    print("Нет уникальных ключевых слов")

```

```python
# 3. Различия - ключевые слова, которые есть только во второй таблице
only_in_table_minus = pd.merge(table_minus, table_plus[['Топик']], on='Топик', how='left', indicator=True)
only_in_table_minus = only_in_table_minus[only_in_table_minus['_merge'] == 'left_only'].drop('_merge', axis=1)

print("\nB) Ключевые слова, которые есть ТОЛЬКО во второй таблице:")
print(f"Количество: {len(only_in_table_minus)}")

if len(only_in_table_minus) > 0:
    display(only_in_table_minus)
else:
    print("Нет уникальных ключевых слов")
```

```python
# переименовать колонки в only_in_table_minus и only_in_table_plus

only_in_table_plus_ready = only_in_table_plus.rename(columns={
    'Количество': 'Количество_table_plus',
    'Процент': 'Процент_table_plus'
})
only_in_table_plus_ready['Количество_table_minus'] = 0
only_in_table_plus_ready['Процент_table_minus'] = 0


only_in_table_minus_ready = only_in_table_minus.rename(columns={
    'Количество': 'Количество_table_minus',
    'Процент': 'Процент_table_minus'
})
only_in_table_minus_ready['Количество_table_plus'] = 0
only_in_table_minus_ready['Процент_table_plus'] = 0


general_table = pd.concat([
    common_keywords,
    only_in_table_plus_ready,
    only_in_table_minus_ready
], ignore_index=True)

display(general_table)
```

```python
general_table.to_excel('general_table.xlsx', index=False)
```

```python

# Добавляем флаги
general_table['has_plus'] = (general_table['Количество_table_plus'] > 0) | (general_table['Количество_table_plus'] > 0)
general_table['has_minus'] = (general_table['Количество_table_minus'] > 0) | (general_table['Количество_table_minus'] > 0)

# Определяем порядок сортировки
conditions = [
    general_table['has_plus'] & general_table['has_minus'],  # 1. Оба значения
    general_table['has_plus'] & ~general_table['has_minus'], # 2. Только плюсы
    ~general_table['has_plus'] & general_table['has_minus'], # 3. Только минусы
    ~general_table['has_plus'] & ~general_table['has_minus'] # 4. Все нули
]

general_table['sort_order'] = np.select(conditions, [1, 2, 3, 4], default=4)

# Сортируем
general_table_sorted = general_table.sort_values(['sort_order', 'Топик']).reset_index(drop=True)

# Удаляем вспомогательные колонки
general_table_sorted = general_table_sorted.drop(['has_plus', 'has_minus', 'sort_order'], axis=1)

print(general_table_sorted)
```

```python
general_table_sorted['Всего отзывов'] = general_table_sorted['Количество_table_plus'] + general_table_sorted['Количество_table_minus']
general_table_sorted['Динамика'] = general_table_sorted['Количество_table_plus'] - general_table_sorted['Количество_table_minus']


print(general_table_sorted)
```
